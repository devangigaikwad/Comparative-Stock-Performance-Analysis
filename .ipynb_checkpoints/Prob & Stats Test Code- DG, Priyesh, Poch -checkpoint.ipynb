```python
import datetime
import numpy as np
import pandas as pd
%matplotlib inline
import pandas_datareader.data as web
# import fix_yahoo_finance as yf
from pandas_datareader import data as web
import yfinance as yf
from scipy import stats
import matplotlib.pyplot as plt
```

```python
symbols = ['HSBA.L','NWG.L','TSCO.L','BARC.L', 'LLOY.L']
```

```python
raw_data = yf.download(symbols, start = '2016-06-23', end='2016-8-23') #start date is 2013/1/1 and end date is 2013/1/2
```

```python
raw_data.head()
```

```python
raw_data['Close'].plot()
```

```python
log_return = np.log(raw_data['Close'] / raw_data['Open'])
```

```python
log_return.hist(bins = 50, figsize = (10, 10))
```

```python
ax1 = plt.subplot(221)
# x = stats.t.rvs(3, size=nsample, random_state=rng)
res = stats.probplot(log_return['BARC.L'], plot=plt)
```

```python
fig, axs = plt.subplots(ncols=2, nrows=3, figsize=(10, 10))

for i, sym in enumerate(symbols):
    
    r = i//2
    c = i%2
    
    res = stats.probplot(log_return[sym], plot=axs[r, c])
    axs[r, c].set_title(sym)
```
